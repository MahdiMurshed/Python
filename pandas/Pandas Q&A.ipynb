# 14.When should you use groupby

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.agg(['count','min','max','mean'])
```

```python
drinks.beer_servings.agg(['count','min','max','mean'])
```

```python
drinks.groupby('continent').agg(['count','min','max','mean'])
```

```python
%matplotlib inline
```

```python
drinks.groupby('continent').mean().plot(kind='bar')
```

```python
ufo.head()
```

```python
ufo.isnull().head()
```

```python
ufo.isnull().sum()
```

```python
ufo.isnull().sum(axis=1).head()
```

```python
ufo[ufo.City.isnull()]
```

```python
ufo.shape
```

```python
ufo.dropna().shape #ufo.dropna(how='any',inplace=False)
```

```python
ufo.dropna(how='all').shape
```

```python
ufo.dropna(subset=['City','Shape Reported'],how='any').shape
```

```python
ufo.dropna(subset=['City','Shape Reported'],how='all').shape
```

```python
ufo['Shape Reported'].value_counts(dropna=False)
```

```python
ufo['Shape Reported'].fillna(value='VARIOUS').value_counts()
```
